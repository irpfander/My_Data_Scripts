# Histograms with Counts and Percents

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
# Set directory for figures
fig_dir = r''

# Read data
df = pd.read_csv(r'')
```

### Choosing color palette

```python
palette_name = 'Set2'
sns.color_palette("Set2", 8)
```

```python
# Plot histogram
palette = sns.color_palette(palette_name, 3)

fig, ax1 = plt.subplots(figsize=(10,7))
graph = sns.countplot(ax=ax1,x='Redacted', data=df, palette=[palette[2],palette[0],palette[1]], hue='Redacted')
graph.set_xticklabels(graph.get_xticklabels(),rotation=45)

plt.xlabel("Redacted")
plt.ylabel("Redacted")
plt.tight_layout()

i=0
for p in graph.patches:
    height = p.get_height()
    graph.text(p.get_x()+p.get_width()/2., height + 0.1,
        height,ha="center")
    i += 1

legend = ax1.get_legend()
handles = legend.legendHandles
ax1.legend(handles, ['Redacted', 'Redacted', 'Redacted'], title='Redacted')
```

```python
palette = sns.color_palette(palette_name, 2)

# filter out no data
df_w_data = df[df.Redacted !='No Data']

fig, ax1 = plt.subplots(figsize=(7,5))
graph = sns.histplot(ax=ax1,x='Redacted', data=df_w_data, palette=palette, hue='Redacted', multiple = "stack")
plt.setp(graph.get_xticklabels(), rotation=45)

[p.get_height() for p in graph.patches]
i=0
for p in graph.patches[:8]:
    height = p.get_height()
    height2 = graph.patches[i+8].get_height()
    graph.text(p.get_x()+p.get_width()/2., height2+height + 0.1,
        "{:.2f}".format((height/(height +height2))*100)+'%',ha="center")
    i += 1
    
plt.xlabel("Redacted")
plt.ylabel("Redacted")
ax1.legend(handles, ['Redacted', 'Redacted', 'Redacted'], title='Redacted')
plt.tight_layout()
```
